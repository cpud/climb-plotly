```python
#Trying to get number of crags on state page.

import requests
from bs4 import BeautifulSoup

URL = 'https://www.mountainproject.com/area/105905173/alabama'
page = requests.get(URL)
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
results = soup.find(id = 'climb-area-page')
```

```python
print(results.prettify())
```

```python
link = results.find('a')['href']
```

```python
print(link)
```

```python
area_elems = results.find_all('section', class_="max-height max-height-md-0 max-height-xs-150")
```

```python
for area_elem in area_elems:
    print(area_elem)
```
